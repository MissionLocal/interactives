```python
# filter out the intersections
df_popup_TL17_SUTTER_ST = df_popup[df_popup['primary_rd'].isin(['SUTTER ST']) & df_popup['secondary_rd'].isin(['POLK ST', 'LARKIN ST', 'HYDE ST', 'LEAVENWORTH ST', 'JONES ST', 'TAYLOR ST', 'MASON ST'])]
df_popup_TL17_POST_ST = df_popup[df_popup['primary_rd'].isin(['POST ST']) & df_popup['secondary_rd'].isin(['POLK ST', 'LARKIN ST', 'HYDE ST', 'LEAVENWORTH ST', 'JONES ST', 'TAYLOR ST', 'MASON ST'])]
df_popup_TL17_GEARY_ST = df_popup[df_popup['primary_rd'].isin(['GEARY ST']) & df_popup['secondary_rd'].isin(['POLK ST', 'LARKIN ST', 'HYDE ST', 'LEAVENWORTH ST', 'JONES ST', 'TAYLOR ST', 'MASON ST'])]
df_popup_TL17_OFARRELL_ST = df_popup[df_popup['primary_rd'].isin(['OFARRELL ST']) & df_popup['secondary_rd'].isin(['POLK ST', 'LARKIN ST', 'HYDE ST', 'LEAVENWORTH ST', 'JONES ST', 'TAYLOR ST', 'MASON ST'])]
df_popup_TL17_ELLIS_ST = df_popup[df_popup['primary_rd'].isin(['ELLIS ST']) & df_popup['secondary_rd'].isin(['POLK ST', 'LARKIN ST', 'HYDE ST', 'LEAVENWORTH ST', 'JONES ST', 'TAYLOR ST', 'MASON ST'])]
df_popup_TL17_EDDY_ST = df_popup[df_popup['primary_rd'].isin(['EDDY ST']) & df_popup['secondary_rd'].isin(['POLK ST', 'LARKIN ST', 'HYDE ST', 'LEAVENWORTH ST', 'JONES ST', 'TAYLOR ST', 'MASON ST'])]
df_popup_TL17_TURK_ST = df_popup[df_popup['primary_rd'].isin(['TURK ST']) & df_popup['secondary_rd'].isin(['POLK ST', 'LARKIN ST', 'HYDE ST', 'LEAVENWORTH ST', 'JONES ST', 'TAYLOR ST','MASON ST'])]
df_popup_TL17_GOLDEN_GATE_AVE = df_popup[df_popup['primary_rd'].isin(['GOLDEN GATE AVE']) & df_popup['secondary_rd'].isin(['POLK ST', 'LARKIN ST', 'HYDE ST', 'LEAVENWORTH ST', 'JONES ST', 'TAYLOR ST'])]
df_popup_TL17_MCALLISTER_ST = df_popup[df_popup['primary_rd'].isin(['MCALLISTER ST']) & df_popup['secondary_rd'].isin(['POLK ST', 'LARKIN ST', 'HYDE ST', 'LEAVENWORTH ST', 'JONES ST', '7TH ST'])]
df_popup_TL17_FULTON_ST = df_popup[df_popup['primary_rd'].isin(['FULTON ST']) & df_popup['secondary_rd'].isin(['LARKIN ST', 'HYDE ST'])]
df_popup_TL17_GROVE_ST = df_popup[df_popup['primary_rd'].isin(['GROVE ST']) & df_popup['secondary_rd'].isin(['POLK ST','LARKIN ST', 'HYDE ST'])]
df_popup_TL17_VAN_NESS_AVE = df_popup[df_popup['secondary_rd'].isin(['VAN NESS AVE']) & df_popup['primary_rd'].isin(['SUTTER ST', 'POST ST', 'GEARY ST', 'OFARRELL ST', 'ELLIS ST', 'EDDY ST', 'TURK ST', 'GOLDEN GATE AVE', 'MCALLISTER ST', 'GROVE ST'])]
df_popup_TL17 = pd.concat([df_popup_TL17_SUTTER_ST, df_popup_TL17_POST_ST, df_popup_TL17_GEARY_ST, df_popup_TL17_OFARRELL_ST, df_popup_TL17_ELLIS_ST, df_popup_TL17_EDDY_ST, df_popup_TL17_TURK_ST, df_popup_TL17_GOLDEN_GATE_AVE, df_popup_TL17_MCALLISTER_ST, df_popup_TL17_FULTON_ST, df_popup_TL17_GROVE_ST, df_popup_TL17_VAN_NESS_AVE])
df_popup_TL17
```

```python
# get a new column that is just date from the df_popup_TL17
df_popup_TL17['collision_date'] = pd.to_datetime(df_popup_TL17['collision_datetime']).dt.date
# change the collision_date column to a string
df_popup_TL17['collision_date_str'] = df_popup_TL17['collision_date'].astype(str)
# get a new column that date from collision_date in the formate of January 1, 2021
df_popup_TL17['collision_date_full'] = pd.to_datetime(df_popup_TL17['collision_date']).dt.strftime('%B %d, %Y')
```

```python
# drop the collision_datetime column
df_popup_TL17 = df_popup_TL17.drop(columns=['collision_datetime'])
# drop the collision_date column
df_popup_TL17 = df_popup_TL17.drop(columns=['collision_date'])
```

```python
# add a new column so that: 
# 'Pedestrian' 'Passenger' 'Parked Vehicle' 'Other' 'Driver' 'Bicyclist'
#if victim_role is Pedestrian, then it is Pedestrian/Bicyclist,
#if victim_role is Passenger, then it is Car person;
#if victim_role is Parked Vehicle, then it is Car person;
#if victim_role is Other, then it is Other;
#if victim_role is Driver, then it is Car person; 
#if victim_role is Bicyclist, then it is Pedestrian/Bicyclist;

df_popup_TL17['role'] = df_popup_TL17['victim_role'].apply(lambda x: 'Pedestrian/Bicyclist' if x == 'Pedestrian' or x == 'Bicyclist' else 'Car person' if x == 'Passenger' or x == 'Parked Vehicle' or x == 'Driver' else 'Other')
df_popup_TL17.head(1)
```

```python
#get the type of each column
df_popup_TL17.dtypes
```

```python
# turn numbr_killed and number_injured into integers
df_popup_TL17['number_killed'] = df_popup_TL17['number_killed'].astype(int)
df_popup_TL17['number_injured'] = df_popup_TL17['number_injured'].astype(int)
```

```python
import geopandas as gpd
from shapely.geometry import Point

# create a GeoDataFrame
gdf = gpd.GeoDataFrame(df_popup_TL17, geometry=gpd.points_from_xy(df_popup_TL17.tb_longitude, df_popup_TL17.tb_latitude))

# set the coordinate reference system (CRS) to EPSG 4326
gdf.crs = {'init': 'epsg:4326'}
```

```python
# split the gdf into 6 dataframes based on accident_year
gdf_2018 = gdf[gdf['accident_year'] == 2018]
gdf_2019 = gdf[gdf['accident_year'] == 2019]
gdf_2020 = gdf[gdf['accident_year'] == 2020]
gdf_2021 = gdf[gdf['accident_year'] == 2021]
gdf_2022 = gdf[gdf['accident_year'] == 2022]
gdf_2023 = gdf[gdf['accident_year'] == 2023]
```

```python
# export the six new dataframes to geojson
gdf_2018.to_file("TL2018.geojson", driver='GeoJSON')
gdf_2019.to_file("TL2019.geojson", driver='GeoJSON')
gdf_2020.to_file("TL2020.geojson", driver='GeoJSON')
gdf_2021.to_file("TL2021.geojson", driver='GeoJSON')
gdf_2022.to_file("TL2022.geojson", driver='GeoJSON')
gdf_2023.to_file("TL2023.geojson", driver='GeoJSON')
```
