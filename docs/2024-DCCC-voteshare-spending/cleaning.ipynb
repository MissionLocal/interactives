## Merge data

```python
df
```

```python
votes = pd.read_csv('votes.csv')
```

```python
votes
```

```python
# group by contest and sum the votes but also keep the slate 

df = df.groupby(["contest", "slate"]).sum().reset_index()
```

```python
df
```

```python
# rename the amount column to amount_raised 

df = df.rename(columns={"amount": "amount_raised"})
```

```python
# delete the row where the contest is multiple candidates RCCC

df = df[df.contest != "Multiple Candidates Rccc"]
```

```python
df
```

```python
# make a list of all the contests

contests = df.contest.unique()
```

```python
contests
```

```python
votes.contest.unique()
```

```python
votes.columns
```

```python
# change amount to number_votes 

votes = votes.rename(columns={"amount": "number_votes"})
```

```python
votes
```

```python
# in df["contest"] replace Sandra Lee Fewer with Sandra Lee Fewer for Supervisor

df.loc[(df["contest"] == "Sandra Fewer"), "contest"] = "Sandra Lee Fewer"
df.loc[(df["contest"] == "Albert Zecher"), "contest"] = "Chip Zecher"
df.loc[(df["contest"] == "Peter Lee"), "contest"] = "Peter Ho Lik Lee"
df.loc[(df["contest"] == "Laurance Lee"), "contest"] = "Laurance Lem Lee"
```

```python
# join the two dataframes on contest

df = df.merge(votes, on="contest", how='outer')
```

```python
df
```

```python
# only keep the rows where the race is DCCC 

df = df[df['race'] == 'DCCC']
```

```python
# show column types

df.dtypes
```

```python
# if amount raised is null, replace with 0

df["amount_raised"] = df["amount_raised"].fillna(0)
```

```python
df.dtypes
```

```python
# save as csv

df.to_csv('all_data.csv', index=False)
```
